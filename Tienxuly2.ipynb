Tiền xử lý ảnh: resize


```python
# hàm tiền xử lý ảnh
from PIL import Image
def resize_images(input_folder, output_folder, size):
    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Get the list of files in the input folder
    files = os.listdir(input_folder)

    for file_name in files:
        # Generate the full path of the input and output files
        input_path = os.path.join(input_folder, file_name)
        output_path = os.path.join(output_folder, file_name)

        # Open the image using PIL
        image = Image.open(input_path)

        resized_image = image.resize(size)

        # Save the resized image to the output folder
        resized_image.save(output_path)
size = (224, 224)
# Example usage
root_dir = "/content/drive/MyDrive/XLA/Leaves_fix/"
for input_foder in os.listdir(root_dir):
    input="/content/drive/MyDrive/XLA/Leaves_fix/" + input_foder
    resize_images(input, input, size)
```

Trích chọn đặc trưng bằng HOG


```python
source_folder1_path = "/content/drive/MyDrive/XLA/Leaves_fix/"

# Đường dẫn đến thư mục chứa các thư mục con
destination_folder1_path = "/content/drive/MyDrive/XLA/Features/"

# Lấy danh sách tất cả các folder trong thư mục gốc
folders = os.listdir(source_folder1_path)

# Sắp xếp các folder theo thứ tự tên
sorted_folders = sorted(folders)


for i, folder_name in enumerate(sorted_folders):

    # Đường dẫn đến folders nguồn
    source_folder2_path = os.path.join(source_folder1_path,folder_name)

    # Đường dẫn đến folders đích
    destination_folder2_path = os.path.join(destination_folder1_path, folder_name)

    # Lấy danh sách tất cả các files trong thư mục gốc
    files = os.listdir(source_folder2_path)

    # Sắp xếp các files theo thứ tự tên
    sorted_files = sorted(files)


    for i, file_name in enumerate(sorted_files):

        # Đường dẫn đến file nguồn
        source_file_path = os.path.join(source_folder2_path, file_name)

        # Đường dẫn đến file đích
        destination_file_path = os.path.join(destination_folder2_path, file_name)

        # Load the image,ảnh xám
        img = cv2.imread(source_file_path,0)

        # cv2_imshow(img)
        H = feature.hog(img, orientations=9, pixels_per_cell=(8, 8),
                        cells_per_block=(2, 2), transform_sqrt=True, block_norm="L2")


        # print('Kích thước hog features: ', H.shape)
        (H, hogImage) = feature.hog(img, orientations=9, pixels_per_cell=(8, 8),
            cells_per_block=(2, 2), transform_sqrt=True, block_norm="L2",
            visualize=True)

        hogImage = exposure.rescale_intensity(hogImage, out_range=(0, 255))
        hogImage = hogImage.astype("uint8")

        cv2.imwrite(destination_file_path, hogImage)

        #plt.imshow(hogImage)
```
